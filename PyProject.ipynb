```python
from bs4 import BeautifulSoup
import requests

res=requests.get(f"https://www.worldometers.info/coronavirus/","htmlparser")
data=res.text
soup=BeautifulSoup(data,'lxml')
#soup
tags=soup.table
tag=tags.get_text()
tag.split()

tags=soup.table
tag=tags.find_all('tr',align=False)
#tag[0].get_text()
final_data=[]
tag=tag[0:8]
for i in tag:
    final_data.append(i.get_text())
# final_data[1].split(',')
data1=[]
for x in range(len(final_data)):
    data1.append(final_data[x][1:-1].split('\n'))
data1
```

```python
import csv
len_list=len(data1)
with open('corona2.csv','a') as csvFile:
    for i in range(len_list):
        writer=csv.writer(csvFile)
        writer.writerow(data1[i])
        
```
