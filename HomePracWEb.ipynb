```python
from bs4 import BeautifulSoup
import requests

#class Scrap:
n=int(input('enter value of n= '))
finaldata=[]
for i in range(1,n+1):
    res=requests.get(f'http://www.nepalstock.com/main/floorsheet/index/{i}/','html.parser')
    data=res.text
    soup=BeautifulSoup(data,'lxml')
    #soup
    tags=soup.table
    tag=tags.find_all('tr',align=False)
#     tag=tags.get_text()
#     tag.split()
    tag=tag[1:-3]
    final_data=[]
    for x in tag:
        final_data.append(x.get_text())
    
    for y in range(len(final_data)):
        finaldata.append(final_data[y][1:-1].split('\n'))
finaldata
    
```

```python
import csv
list_len=len(finaldata)
with open('scrap.csv','a') as csvFile:
    for i in range(list_len):
        writer=csv.writer(csvFile)
        writer.writerow(finaldata[i])
```

```python
from bs4 import BeautifulSoup
import requests

store1=[]
n=int(input("Enter no. of page= "))
for i in range(1,n+1):
    res=requests.get(f"http://www.nepalstock.com/main/floorsheet/index/{i}/","htmlparser")
    data=res.text
    soup=BeautifulSoup(data,'lxml')
    tags=soup.table
    tag=tags.find_all('tr',align=False)
    req_data=[]
    for x in tag:
        req_data.append(x.get_text())
# req_data[0].split('\n')
    for y in range(len(req_data)):
        store1.append(req_data[y][1:-1].split('\n'))
store1        
```

```python
import csv
list_len=len(store1)
with open("homePrac1.csv","a") as csvFile:
    for m in range(list_len):
        write=csv.writer(csvFile)
        write.writerow(store1[m])
```
