```python
import requests
from bs4 import BeautifulSoup

total=[]
n=int(input("Enter page you want: "))
for page in range(1,n+1):
    res=requests.get(f'http://www.nepalstock.com/main/floorsheet/index/{page}/','html.parser')
    content=res.text
    soup=BeautifulSoup(content,'lxml')
    # tags=soup.table
    # tag=tags.get_text()
    # tag.split()
    # tag=tags.find_all('tr',align=False)
    # tag
    tags=soup.table
    tag=tags.find_all('tr',align=False)

    all_data=[]
    tag=tag[0:-2]
    #contents[0].get_text().split('\n')
    for j in tag:
        all_data.append(j.get_text())
    #all_data  it import in form like \n1\n2020051301000206\nADBL\n16\n45\n100\n347\n34700.00\n',  
    for k in range(len(all_data)):
        total.append(all_data[k][1:-1].split('\n'))
total
    
```

```python
import csv

lenlist=len(total)
with open('pythonPrac1.csv','w+') as csvFile:
    for i in range(lenlist):
        w=csv.writer(csvFile)
        w.writerow(total[i])
```
