```python
from bs4 import BeautifulSoup
import requests

final_data=[]
res=requests.get(f"https://www.worldometers.info/coronavirus/","htmlparser")
contents=res.text
soup=BeautifulSoup(contents,'lxml')
tags=soup.table
tag=tags.find_all('tr')
#tag[0].get_text().split('\n')
data=[]
tag=tag[0:-1]
for i in tag:
    data.append(i.get_text())
data[0].split('\n')
for y in range(len(data)):
    final_data.append(data[y][1:-1].split('\n'))
final_data
```

```python
import csv

list_ln=len(final_data)
with open("wprldmeter.csv","a") as csvFile:
    for m in range(list_ln):
        writerr=csv.writer(csvFile)
        writerr.writerow(final_data[m])
        
```
